```python
def yellow_line_region_trend():
    """
    畫折線圖：比較 North 和 South 兩個地區的月營收趨勢
    - X 軸：月份
    - Y 軸：該月總營收
    - 兩條線，有圖例 (legend)
    回傳 matplotlib Figure 物件
    提示：分別 groupby 再 plot，或用 sns.lineplot(hue='region')
    """
    # TODO: 你的程式碼
target_regions = df[df['region'].isin(['North', 'South'])].copy()
target_regions['month'] = target_regions['order_date'].dt.month

region_rev = target_regions.groupby(['month','region'])['amount'].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=region_rev, x='month', y='amount', hue = 'region', marker='o', linewidth=2)
plt.title('Monthly region Trend', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('total revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def yellow_box_vip():
    """
    畫箱形圖：比較不同 VIP 等級 (vip_level) 的訂單金額分佈
    回傳 matplotlib Figure 物件
    提示：sns.boxplot(x='vip_level', y='amount', data=df)
    """
    # TODO: 你的程式碼
plt.figure(figsize=(9, 5))
sns.boxplot(data=df, x='vip_level', y='amount', palette='Set2', hue='vip_level', legend=False)
plt.title('Order Amount Distribution by vip level', fontweight='bold')
plt.xlabel('vip_level')
plt.ylabel('Amount')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
def yellow_scatter_price_amount():
    """
    畫散佈圖：X=商品單價 (unit_price)，Y=訂單金額 (amount)
    回傳 matplotlib Figure 物件
    提示：plt.scatter() 或 sns.scatterplot()
    """
    # TODO: 你的程式碼
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='unit_price', y='amount',
                hue='category', alpha=0.6, s=60)
plt.title('Unit Price vs Order Amount (by Category)', fontweight='bold')
plt.xlabel('Unit Price')
plt.ylabel('Order Amount')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show() 
```

```python
def red_category_dashboard(category="Electronics"):
    """
    針對指定類別，畫 2×2 的 subplot dashboard：
    1. 左上：該類別月營收趨勢 (折線圖)
    2. 右上：該類別各地區營收 (長條圖)
    3. 左下：該類別 Top 5 商品營收 (水平長條圖)
    4. 右下：該類別訂單金額分佈 (直方圖)

    回傳 matplotlib Figure 物件
    提示：fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    """
    # TODO: 你的程式碼
fig, axes = plt.subplots(2, 2, figsize=(16, 9))
fig.suptitle('Electronics Sales Dashboard — 2025', fontsize=16, fontweight='bold')

# (0,0) 月度趨勢
electronics_df = df[df['category'].str.lower() == 'electronics']
region_electronics_rev = electronics_df.groupby('region')['amount'].sum().reset_index()
plt.figure(figsize=(8, 4)) 
sns.lineplot(data=region_electronics_rev, x='region', y='amount', marker='o', ax=axes[0, 0])
plt.title('Electronics Revenue by Region', fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Revenue (NT$)')

# (0,1) 地區排名
sns.barplot(data=region_electronics_rev, x='region', y='amount', ax=axes[0, 1],
            palette='viridis', hue='region', legend=False)
axes[0, 1].set_title('electronics_Revenue by Region')

# (1,0) 品類分布

electronics_df = df[df['category'].str.lower() == 'electronics']
product_rev = electronics_df.groupby('product_name')['amount'].sum().reset_index()
top5_products = product_rev.sort_values('amount', ascending=False).head(5)
fig = plt.figure(figsize=(10, 6))
sns.barplot(
        data=top5_products, 
        x='amount',          
        y='product_name',     
        hue='product_name', 
        legend=False
    )

# (1,1) 金額散佈
sns.scatterplot(data=df, x='category', y='amount',
                hue='category', alpha=0.6, ax=axes[1, 1], legend=False)
axes[1, 1].set_title('electronics Amount')

plt.tight_layout()
plt.show()
```
