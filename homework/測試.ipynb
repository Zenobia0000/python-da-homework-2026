```python
"""
M4 時間序列與 EDA — 課後作業
==============================
情境：用合併好的訂單資料做時間維度分析，
產出月報級別的商業洞察。

資料路徑：datasets/ecommerce/orders_enriched.csv
"""
import pandas as pd


def _load_data():
    """輔助函式：使用絕對路徑讀取並解析日期"""
    # 直接指向檔案在電腦裡的完整位置
    path = r"C:\Users\W11 HOME\Desktop\python-da-homework-2026\datasets\ecommerce\orders_enriched.csv"
    df = pd.read_csv(path, parse_dates=["order_date"])
    return df


df = _load_data()
print(_load_data().head(2))  # 顯示前 5 筆資料


# ============================================================
# 🟢 送分題（每題 10 分，共 30 分）
# ============================================================

def green_top3_dates(df):
    """
    找出訂單數最多的前 3 個日期
    回傳 Series（index=日期, values=訂單數, 由多到少排序）
    提示：value_counts().head(3)
    """
    # 1. 確保 order_date 欄位已轉換為 datetime 格式
    df['order_date'] = pd.to_datetime(df['order_date'])
    
    # 2. 使用 value_counts() 計算每個日期的出現次數（即訂單數）
    # 3. 使用 head(3) 取得前三名並回傳
    return df['order_date'].value_counts().head(3)

# --- 測試執行 (請確保這兩行在最左邊，不要縮排) ---
df = _load_data() 
print(green_top3_dates(df))
```
